```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
t = np.arange(0., 2., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'rs', t, t**3, 'g^')
plt.show()
```

```python
import pandas as pd
# Create a csv file for student result having student name, marks ,percentage.
#read file and plot marks and percentage.
res={'name':['Raj','Ravi', 'Krish'],
     'Python':[70,85,58],
     'java':[60,80,45],
     'cpp':[90,58,85]
    }
```

```python
data=pd.DataFrame(res)
data
```

```python
data['total']= data['Python']+data['java']+data['cpp']
data
```

```python
data['percentage']= data['total']/3
data
```
